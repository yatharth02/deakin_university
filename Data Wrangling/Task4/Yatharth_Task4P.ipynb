## Introduction
In this case study we will be learning the usage of pandas with help of the Airline on-time data for all flights departing from NYC in 2013. This dataset contains different features i.e.:
- `origin` –> weather station:LGA, JFK or EWR.
- `year`, `month`, `day`, `hour` -> time of recording.
- `temp`, `dewp` –> temperature and dew point in degrees Fahrenheit.
- `humid` –> relative humidity
- `wind_dir`, `wind_speed`, `wind_gust` –> wind direction(in degrees), speed and gust speed(in mph)
- `precip` –> precipitation in inches
- `pressure` –> sea level pressure in millibars
- `visib` –> visibility in miles
- `time_hour` –> date and hour(based on the year, month, day,hour fields)formatted as `YYYY-mm-dd HH:MM:SS`.

Defining imports at the top that will be used by the report.
In this case, having all the imports at the top is more intuitive than having some of them at the top and some of them scattered over the file.

```python
import warnings

import numpy as np
import pandas as pd

# Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
import matplotlib.pyplot as plt

# Seaborn is a Python visualization library based on matplotlib. It provides a high-level interface for drawing attractive statistical graphics.
import seaborn as sns

# Ignore all warnings generated by the Python program
warnings.filterwarnings("ignore")

# This command tells the IPython environment to draw the plots immediately after the current cell
%matplotlib inline
```

Extracting data into pandas Dataframe from a compressed csv file, ignoring comments.

```python
ny_flight_df = pd.read_csv('nycflights13_weather.csv.gz', comment='#')

# Printing top 5 data from our dataframe
ny_flight_df.head()
```

```python
print(f'Flight dataset have {ny_flight_df.shape[0]} entries for {ny_flight_df.shape[1]} features.')
```

Due to a bug in the dataset, the data in the `time_hour` column are incorrectly shifted by 1 hour, so rectifying the dataset.

```python
# Converting time_hour column dtype from object to datetime64.
ny_flight_df.time_hour = ny_flight_df.time_hour.astype('datetime64[ns]')

# Subtracting 1 hour from dataset.
ny_flight_df.time_hour = ny_flight_df.time_hour - pd.to_timedelta(1, unit='h')
```

Converting columns so that they can use metric(International System of Units, SI) or derived units.

```python
# Converting columns temp and dewp, temperature and dew point in degrees Fahrenheit to degrees Celsius.

def fahrenheit_to_celsius(val):
    """
    Converting degrees Fahrenheit to degrees Celsius.
    :param val: Value in degrees Fahrenheit.
    :return: Value in degrees Celsius.
    """
    return (val - 32) * 5 / 9


ny_flight_df[['temp', 'dewp']] = ny_flight_df[['temp', 'dewp']].apply(fahrenheit_to_celsius)
```

```python
# Converting column precip, precipitation in inches to millimetres.
ny_flight_df.precip = ny_flight_df.precip * 25.4
```

```python
# Converting column visib, visibility in miles to metres
ny_flight_df.visib = ny_flight_df.visib * 1609.34
```

```python
# Converting columns wind_speed and wind_gust, wind speed and wind gust speed in mph to mps
ny_flight_df[['wind_speed', 'wind_gust']] = ny_flight_df[['wind_speed', 'wind_gust']] * 0.44704
```

As we have converted columns to generic units, now we will be using this dataset for our further analysis.

Now from the converted dataset we will be using LGA airport from the `origin` column and will be calculating daily mean of `wind_speed`.

```python
# Creating a new dataframe with data only with LGA airport as origin and daily mean of wind_speed
lga_flight_wind_mean_df = (ny_flight_df[ny_flight_df.origin == 'LGA']
                           .groupby(['year', 'month', 'day'])['wind_speed']
                           .mean()
                           .reset_index(drop=False))

# Creating column date for further analysis.
lga_flight_wind_mean_df['date'] = pd.to_datetime(
    dict(year=lga_flight_wind_mean_df.year, month=lga_flight_wind_mean_df.month, day=lga_flight_wind_mean_df.day)
)

print(
    f'Daily mean of wind speeds for the LGA airport dataset have {lga_flight_wind_mean_df.shape[0]} entries for {lga_flight_wind_mean_df.shape[1]} features.')
```

As we have created a new dataset of LGA airport with daily mean, now we will be plotting a graph using line plot, which is a graph that uses lines to connect data points to display data as points or check marks above a number line, showing the frequency of each value.
Here we will be creating a line plot for our dataset.

```python
# set the figure size
plt.figure(figsize=(15, 6))

# Creating line plot
sns.lineplot(lga_flight_wind_mean_df, x='date', y='wind_speed', color='black')

# Set the x-axis and y-axis labels
plt.xlabel('day')
plt.ylabel('daily average wind speed [m/s] at LGA')

# Will turn off the grid
plt.grid(False)

# Display plot
plt.show()
```

As per the visualization, we can derive:
- In 2013, daily average wind speed ranges from 1.5 to 11.5.
- Very high wind speed may lead to delay in flight.

```python
# Fetching 10 windiest days at LGA airport with dates and the corresponding mean of daily wind speeds.
top_10_lga_flight_wind_mean_df = (lga_flight_wind_mean_df[['date', 'wind_speed']]
                                  .sort_values(by='wind_speed', ascending=False)
                                  .reset_index(drop=True)
                                  .set_index('date')
                                  .head(10))
top_10_lga_flight_wind_mean_df
```

As of now, we need to check about outliers in `wind_speed`, so after which we can strategize our way to handle the outliers.

```python
# set the figure size
plt.figure(figsize=(15, 6))

# Creating box plot
sns.boxplot(ny_flight_df['wind_speed'], orient='h')

# Display plot
plt.show()
```

As per the visualization, we can derive that there are few outliers, but from which, there is 1 outlier which lies far beyond the others. So we can replace this maximum value with NAN, so that our further analysis won't be hampered.

```python
# Replacing the max value which has been checked from boxplot outlier detection
ny_flight_df['wind_speed'].replace(ny_flight_df['wind_speed'].max(), np.NAN, inplace=True)
```

```python
# Creating a new dataframe with monthly mean of wind speeds for all the 3 airports
origin_flight_wind_mean_df = (ny_flight_df.groupby(['origin', 'month', 'year'])['wind_speed']
                              .mean()
                              .reset_index(drop=False))

# Creating column month_year for further analysis.
origin_flight_wind_mean_df['month_year'] = pd.to_datetime(
    origin_flight_wind_mean_df[['year', 'month']].apply(lambda row: '-'.join(row.values.astype(str)), axis=1),
    format='%Y-%m')
```

As we have created a new dataset of monthly mean of wind speeds for all the 3 airports, now we will be plotting a graph using line plot, which is a graph that uses lines to connect data points to display data as points or check marks above a number line, showing the frequency of each value.
Here we will be creating a line plot for our dataset.

```python
# set the figure size
plt.figure(figsize=(15, 6))

# Creating line plot
sns.lineplot(origin_flight_wind_mean_df, x='month_year', y='wind_speed', hue='origin')

# Set the x-axis and y-axis labels
plt.xlabel('month')
plt.ylabel('monthly average wind speed [m/s]')

# Will turn off the grid
plt.grid(False)

# Hiding legend title from plot
plt.legend(title=None)

# Display plot
plt.show()
```

As per the visualization, we can derive:
- `JFK` origin flights have encountered more average wind speed than other airports.
- `EWR` origin flights have encountered less average wind speed than other airports.
