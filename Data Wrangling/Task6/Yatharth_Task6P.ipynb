```python
import sqlite3
import warnings

import numpy as np
import pandas as pd

# Ignore all warnings generated by the Python program
warnings.filterwarnings("ignore")
```

Extracting dataset for airlines, which contains carrier codes for airline names.

```python
ny_flight_airlines_df = pd.read_csv('nycflights13_airlines.csv.gz', comment='#')
ny_flight_airlines_df.head()
```

Extracting dataset for airports, which contains information about airports.

```python
ny_flight_airports_df = pd.read_csv('nycflights13_airports.csv.gz', comment='#')
ny_flight_airports_df.head()
```

Extracting dataset for flights, which contains ontime data for all flights that departed from NYC in 2013.

```python
ny_flight_flights_df = pd.read_csv('nycflights13_flights.csv.gz', comment='#')
ny_flight_flights_df.head()
```

Extracting dataset for planes, which contains information about planes found in FAA aircraft registry.

```python
ny_flight_planes_df = pd.read_csv('nycflights13_planes.csv.gz', comment='#')
ny_flight_planes_df.head()
```

Extracting dataset for weather, which contains hourly meteorological data LGA, JFK and EWR.

```python
ny_flight_weather_df = pd.read_csv('nycflights13_weather.csv.gz', comment='#')
ny_flight_weather_df.head()
```

```python
# Opens a connection to the SQLite database file
conn = sqlite3.connect("nycflight.db")

# Exporting dataframes to the database
ny_flight_airlines_df.to_sql("airlines", conn, index=False)
ny_flight_airports_df.to_sql("airports", conn, index=False)
ny_flight_flights_df.to_sql("flights", conn, index=False)
ny_flight_planes_df.to_sql("planes", conn, index=False)
ny_flight_weather_df.to_sql("weather", conn, index=False)
```

### In the task below we will extract unique `engine` data from the plane's dataset.

```python
# Printing dataframe via sql extraction
task1_sql = pd.read_sql_query("""
SELECT DISTINCT engine FROM planes
""", conn)
task1_sql.head()
```

```python
# Printing dataframe via pandas extraction
task1_pd = (
    pd.DataFrame(ny_flight_planes_df['engine'].drop_duplicates().reset_index(drop=True))
)
task1_pd.head()
```

```python
# Comparison of sql-way and pandas-way
print(pd.testing.assert_frame_equal(task1_sql, task1_pd) is None)
```

### In the task below we will fetch the number of `engine` present in the plane's dataset.

```python
# Printing dataframe via sql extraction
task3_sql = pd.read_sql_query("""
SELECT COUNT(*), engine FROM planes GROUP BY engine
""", conn)
task3_sql.head()
```

### In the task below we will fetch the number of combinations of `engine` and `manufacturer`, with aggregation of `year` as minimum, average and maximum data present in the plane's dataset.

```python
# Printing dataframe via sql extraction
task5_sql = pd.read_sql_query("""
SELECT MIN(year), AVG(year), MAX(year), engine, manufacturer FROM planes
GROUP BY engine, manufacturer
""", conn)
task5_sql.head()
```

### In the task below we will print all flights data from the flights dataset with planes data as `plane_year`, `plane_speed` and `plane_seats` from the planes dataset.

```python
# Printing dataframe via sql extraction
task15_sql = pd.read_sql_query("""
 SELECT
flights.*,
planes.year AS plane_year, planes.speed AS plane_speed, planes.seats AS plane_seats
FROM flights LEFT JOIN planes ON flights.tailnum=planes.tailnum
""", conn)
task15_sql.head()
```

```python
# Printing dataframe via pandas extraction
task15_pd = (
    pd.merge(ny_flight_flights_df, ny_flight_planes_df.iloc[:, [0, 1, 7, 6]], how='left', on='tailnum').rename(
        columns={'year_x': 'year', 'year_y': 'plane_year', 'speed': 'plane_speed', 'seats': 'plane_seats'})
)
task15_pd.head()
```

```python
# Comparison of sql-way and pandas-way
print(pd.testing.assert_frame_equal(task15_sql, task15_pd) is None)
```

### In the below task we will print planes and airlines data from flights and planes dataset, where `carrier` and `tailnum` combination need to be unique in the flight dataset, after which will sort by `tailnum` and `carrier` in ascending order.

```python
# Printing dataframe via sql extraction
task16_sql = pd.read_sql_query("""
 SELECT planes.*, airlines.* FROM
(SELECT DISTINCT carrier, tailnum FROM flights) AS cartail INNER JOIN planes ON cartail.tailnum=planes.tailnum
INNER JOIN airlines ON cartail.carrier=airlines.carrier
""", conn)
task16_sql.head()
```

```python
# Printing dataframe via pandas extraction
task16_pd = (
    pd.merge(pd.merge(ny_flight_planes_df, ny_flight_flights_df[['carrier', 'tailnum']].drop_duplicates(), how='inner',
                      on='tailnum'), ny_flight_airlines_df, how='inner', on='carrier').sort_values(
        by=['tailnum', 'carrier']).reset_index(drop=True)
)
task16_pd.head()
```

```python
# Comparison of sql-way and pandas-way
print(pd.testing.assert_frame_equal(task16_sql, task16_pd) is None)
```

### In the task below we will print all flights data from the flight dataset with average of `temp` as `atemp` and average of `humid` as `ahumid` from the weather table where `origin` is 'EWR'.

```python
# Printing dataframe via sql extraction
task17_sql = pd.read_sql_query("""
 SELECT
flights2.*, atemp, ahumid
FROM (
SELECT * FROM flights WHERE origin='EWR'
) AS flights2 LEFT JOIN (
SELECT
year, month, day, AVG(temp) AS atemp, AVG(humid) AS ahumid
FROM weather
WHERE origin='EWR'
GROUP BY year, month, day
) AS weather2
ON flights2.year=weather2.year AND flights2.month=weather2.month AND flights2.day=weather2.day
""", conn)
task17_sql.head()
```

```python
# Printing dataframe via pandas extraction
weather_df = ny_flight_weather_df[ny_flight_weather_df['origin'] == 'EWR'].groupby(['year', 'month', 'day']).agg(
    atemp=('temp', np.mean), ahumid=('humid', np.mean)).reset_index()

task17_pd = (
    pd.merge(ny_flight_flights_df[ny_flight_flights_df['origin'] == 'EWR'], weather_df, how='left',
             on=['year', 'month', 'day'])
)
task17_pd.head()
```

```python
# Comparison of sql-way and pandas-way
print(pd.testing.assert_frame_equal(task17_sql, task17_pd) is None)
```
